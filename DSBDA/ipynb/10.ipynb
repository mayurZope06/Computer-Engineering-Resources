```python
# ---------------------------------------------------
# Data Visualization III - Iris Dataset
# ---------------------------------------------------

# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt

# ---------------------------------------------------
# Load Dataset
# ---------------------------------------------------

df = pd.read_csv("Iris.csv")

# ---------------------------------------------------
# Display First 5 Rows
# ---------------------------------------------------

print("First 5 Rows of Dataset:\n")
print(df.head())

# ---------------------------------------------------
# Shape of Dataset
# ---------------------------------------------------

print("\nShape of Dataset:")
print(df.shape)

# ---------------------------------------------------
# Feature Names
# ---------------------------------------------------

print("\nFeature Names:\n")
print(df.columns)

# ---------------------------------------------------
# Number of Features
# ---------------------------------------------------

print("\nNumber of Features:")
print(len(df.columns))

# ---------------------------------------------------
# Feature Types
# ---------------------------------------------------

print("\nFeature Types:\n")
print(df.dtypes)

# ---------------------------------------------------
# Statistical Summary
# ---------------------------------------------------

print("\nStatistical Summary:\n")
print(df.describe())

# ---------------------------------------------------
# Histograms
# ---------------------------------------------------

df.hist(figsize=(16, 12))

plt.suptitle("Histogram of Iris Dataset Features")

plt.show()

# ---------------------------------------------------
# Boxplots
# ---------------------------------------------------

df.plot(
    kind='box',
    subplots=True,
    layout=(3, 2),
    figsize=(16, 12)
)

plt.suptitle("Boxplot of Iris Dataset Features")

plt.show()
```
