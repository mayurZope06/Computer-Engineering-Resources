```python
# ---------------------------------------------------
# Data Wrangling II
# ---------------------------------------------------

# 1. Import Libraries
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

# ---------------------------------------------------
# 2. Load Dataset
# ---------------------------------------------------

df = pd.read_csv("data.csv")

# ---------------------------------------------------
# 3. Display First 5 Rows
# ---------------------------------------------------

print("First 5 Rows:\n")
print(df.head())

# ---------------------------------------------------
# 4. Check Missing Values
# ---------------------------------------------------

print("\nMissing Values:\n")
print(df.isnull().sum())

# ---------------------------------------------------
# 5. Handle Missing Values
# ---------------------------------------------------

df['Age'] = df['Age'].fillna(df['Age'].mean())

print("\nMissing Values After Filling:\n")
print(df['Age'].isnull().sum())

# ---------------------------------------------------
# 6. Handle Inconsistencies
# ---------------------------------------------------

df['Sex'] = df['Sex'].str.lower()

print("\nUnique Values in Sex Column:\n")
print(df['Sex'].unique())

# ---------------------------------------------------
# 7. Detect Outliers using IQR
# ---------------------------------------------------

Q1 = df['Age'].quantile(0.25)

Q3 = df['Age'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR

upper_bound = Q3 + 1.5 * IQR

print("\nLower Bound:", lower_bound)
print("Upper Bound:", upper_bound)

# ---------------------------------------------------
# 8. Display Outliers
# ---------------------------------------------------

outliers = df[
    (df['Age'] < lower_bound) |
    (df['Age'] > upper_bound)
]

print("\nOutliers:\n")
print(outliers[['Age']])

# ---------------------------------------------------
# 9. Remove Outliers
# ---------------------------------------------------

df = df[
    (df['Age'] >= lower_bound) &
    (df['Age'] <= upper_bound)
]

print("\nDataset Shape After Removing Outliers:")
print(df.shape)

# ---------------------------------------------------
# 10. Boxplot Before Transformation
# ---------------------------------------------------

sns.boxplot(x=df['Age'])

plt.title("Boxplot of Age")

plt.show()

# ---------------------------------------------------
# 11. Data Transformation
# ---------------------------------------------------

df['Age_log'] = np.log(df['Age'])

print("\nAge Column After Log Transformation:\n")
print(df[['Age', 'Age_log']].head())

# ---------------------------------------------------
# 12. Histogram After Transformation
# ---------------------------------------------------

sns.histplot(df['Age_log'], bins=20)

plt.title("Histogram of Log Transformed Age")

plt.show()

# ---------------------------------------------------
# 13. Final Dataset
# ---------------------------------------------------

print("\nFinal Dataset:\n")
print(df.head())
```
