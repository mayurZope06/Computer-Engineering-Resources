```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import MinMaxScaler

# Load dataset
df = pd.read_csv("Titanic-Dataset.csv")

print("First 5 Rows of Dataset:\n")
print(df.head())

# Dataset information
print("\nDataset Information:\n")
print(df.info())

# Check dimensions
print("Rows and Columns:", df.shape)

# ---------------------------------------------------
# Size of Dataset
# ---------------------------------------------------

print("\nTotal Elements in Dataset:")
print(df.size)

df.isnull()

# Check missing values
print("\nMissing Values:\n")
print(df.isnull().sum())

# ---------------------------------------------------
# Non Missing Values
# ---------------------------------------------------

print("\nNon Missing Values:\n")
print(df.notnull().sum())


df.notnull()

df.notnull().sum()


# Statistical summary
print("\nStatistical Summary:\n")
print(df.describe())

df.size


df.shape

# Data types
print("\nData Types:\n")
print(df.dtypes)

# Convert datatype
df['Age'] = df['Age'].astype(float)

print("\nDatatype of Age Column:")
print(df['Age'].dtype)

# Fills missing values in Age with the average age.
df['Age'] = df['Age'].fillna(df['Age'].mean())
print("\nMissing Values After Filling Age Column:\n")
print(df['Age'].isnull().sum())

# Convert categorical to numeric
df['Sex'] = df['Sex'].map({'Male':1, 'Female':0})
print("\nAfter Converting Sex Column:\n")
print(df[['Sex']].head())
# df["gender"].replace({"female":0,"male":1}, inplace=True)

# ---------------------------------------------------
# Data Normalization
# ---------------------------------------------------

scaler = MinMaxScaler()

df[['Fare']] = scaler.fit_transform(df[['Fare']])

print("\nNormalized Fare Column:\n")
print(df[['Fare']].head())

# ---------------------------------------------------
# Final Dataset
# ---------------------------------------------------

print("\nFinal Dataset:\n")
print(df.head())

```
